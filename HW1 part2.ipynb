```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
train = pd.read_csv('train.csv')
train = train.drop('PassengerId',axis = 1)
train.head()
```

先把性別轉換成數值資料，看一下資料間的相關性，發現性別與存活率相關性最大

```python
train['Sex'] = train['Sex'].map({'male' : 1,'female' : 0}).astype('int')
plt.figure(figsize=(12,10))
ex4 = sns.heatmap(train[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare']].corr(), vmax=0.6, square=True, annot=True)
```

看一下數值資料的敘述統計量
